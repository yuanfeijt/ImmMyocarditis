```python
import scanpy as sc
import os
```

```python
base_dir = '../Data/Mantri_M_et_al/scRNAseq'  # 替换为你的数据目录
sample_dirs = os.listdir(base_dir)  # 替换为你的样本文件夹名
```

```python
adatas = []
for sample_file in sample_dirs:
    # 只处理.h5文件
    if sample_file.endswith('.h5'):
        file_path = os.path.join(base_dir, sample_file)
        if os.path.exists(file_path):
            adata = sc.read_10x_h5(file_path)
            adata.var_names_make_unique()
            # 添加样本标识（使用文件名作为样本ID）
            sample_id = sample_file.replace('_filtered_feature_bc_matrix.h5', '')  # 去掉后缀
            adata.obs['sample'] = sample_id  # 添加样本列

            adatas.append(adata)
            print(f"已成功读取: {sample_file}")
combined_adata = adatas[0].concatenate(
    adatas[1:],
    batch_key="sample",  # 在obs中创建样本标识列
    batch_categories=[adata.obs['sample'].iloc[0] for adata in adatas]  # 样本名称
)
combined_adata.var_names_make_unique()
combined_adata.obs_names_make_unique()
print(f"合并后的数据形状: {combined_adata.shape}")
print(f"样本信息: {combined_adata.obs['sample'].unique().tolist()}")
```

```python
new_gene_names = [gene_name.replace('GRCm38_', '') for gene_name in combined_adata.var_names]
combined_adata.var_names = new_gene_names
print(f"处理后基因名示例: {new_gene_names[:5]}")
```

```python
sc.pp.filter_cells(combined_adata, min_genes=200)
sc.pp.filter_genes(combined_adata, min_cells=3)
print(combined_adata.obs)
sc.pp.scrublet(combined_adata, batch_key="sample")
```

```python
combined_adata.var["mt"] = combined_adata.var_names.str.startswith("mt-")
combined_adata.var["ribo"] = combined_adata.var_names.str.contains("^Mrp[ls]")
combined_adata.var["hb"] = combined_adata.var_names.str.contains("^Hb[^(P)]")
sc.pp.calculate_qc_metrics(
    combined_adata, qc_vars=["mt", "ribo", "hb"], inplace=True, log1p=True
)
combined_adata
```

```python
sc.pl.violin(
    combined_adata,
    ["n_genes_by_counts", "total_counts", "pct_counts_mt"],
    jitter=0.4,
    multi_panel=True,
)
```

```python
combined_adata = combined_adata[combined_adata.obs.predicted_doublet == False, :]
combined_adata = combined_adata[combined_adata.obs.doublet_score <= 0.25, :]
combined_adata
```

```python
combined_adata = combined_adata[combined_adata.obs.n_genes_by_counts < 4000, :]
combined_adata = combined_adata[combined_adata.obs.pct_counts_mt < 20, :]
combined_adata = combined_adata[combined_adata.obs.total_counts < 20000, :]
combined_adata = combined_adata[combined_adata.obs.total_counts > 500, :]
print(combined_adata)
combined_adata.write_h5ad('../Data/Mantri_M_et_al.h5ad')
```
